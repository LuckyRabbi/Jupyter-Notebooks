## 2D Plots, Plotting, Subplots and Labels

Just a quick word about pyplot, this is a state-based interface to matplotlib, it provides a way of plotting that is very similar to MATLAB, however it is not recommended for very complex plots, in which case using the object-oriented API is recommended. 

The most important part of matplotlib pyplot is the `plot` method, this allows one to plot 2D data, the full documentation of plot can be found [here](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot).

The following code shows how to display a plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# First make a range of x and y values for a sine and cosine curve
x = np.arange(-np.pi*2, np.pi*2, 0.1)
def y_sin(x):
    return(np.sin(x))

def y_cos(x):
    return(np.cos(x))

# Now plot the curve
plt.plot(x, y_sin(x), x, y_cos(x)) # plot() takes multiple arguments
plt.show()
```

### Axes, Ticks and Spines

The Axes are very similar to subplots, however they can be placed *anywhere* on the plot, making them a little bit more versatile than subplots, which will be explained later. The [Axis](https://matplotlib.org/users/artists.html#axis-container) container is responsible for handling the drawing of ticks, grid lines, axis labels e.t.c.

Below is an example of how one might change the x-ticks, sine and cosine are much more interesting over values of pi, so I will change the xticks to reflect this, I will also label them using matplotlibs inbuilt text processor built off LaTEX.

#### Changing the ticks

Naturally, a function may be more interesting over a particular set of numbers, for example the plot above shows sine and cosine over a range that is interesting, but we mainly want to look at certain values of pi, so we can make custom x-ticks, in addition, we only really care about maximum and minimum values (or certain values like $\sqrt{2}/2$) so we will change the y-ticks as well. To do this we use the `plt.xticks()` and `plt.yticks()` methods respictively.

You can read more about the [xticks](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xticks) and [yticks](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.yticks) methods to get an understanding of what is happening in the code below. You may also be interested in reading more about [tick formatting](https://matplotlib.org/api/ticker_api.html) as part of the ticker API.

#### Moving the Spines

The sine and cosine graphs are more appropriately graphed going through the origin, by default the spines are on the corners of the graph, we will move them to the center of the graph.

To move the spine we use the `plt.gca()` method, which allows us to manipulate the axes. The spines [documentation](https://matplotlib.org/api/spines_api.html#matplotlib.spines) details how one can manipulate the spines, there is also a really good tutorial on [transforming graphics](https://matplotlib.org/users/transforms_tutorial.html).

The plots have a lot of customisation options, including adding labels to the x and y axis, titles and legends, this is only scratching the surface of the level of customisation, indeed, there are ways to make entirely custom plots using this library.

#### Legends

The legend can be added to a plot by using the `legend()` method, the [documentation](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.legend) should be read before looking at the [legend guide](https://matplotlib.org/users/legend_guide.html) so that you can make some awesome custome legends!

#### Annotations

Annotations are useful especially for resources that are used for teaching and learning, if we want to zoom in on a particular part of the graps that is interesting we can add an annotation to the graph.

I will add a dotted line at $\pi/2$ and also an annotation at this point for the two graphs. I will also add a scatter point at the point to make the label more clear.

To see more about annotations, this is a [guide](https://matplotlib.org/users/annotations_guide.html) and this is the [documentation](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.annotate).

```python
t = np.pi/2 # Point that I want to annotate

plt.xticks([-2*np.pi,-3*np.pi/2,-np.pi, -np.pi/2, 0, np.pi/2, np.pi,3*np.pi/2,2*np.pi],
           [r'$-2\pi$',r'$-3\pi/2$',r'$-\pi$',r'$-\pi/2$',r'$0$',
            r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$1$'])

plt.title("Sine and Cosine") # Adding a title

ax = plt.gca()  # gca stands for 'get current axis' and allows for manipulation
ax.spines['right'].set_color('none') # Discard the top and right axis
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# Plotting individually to add them to the legend
plt.plot(x, y_sin(x), color="blue", label="Sine")
plt.plot(x, y_cos(x), color="red", label="Cosine")

plt.plot([t,t], [0,y_sin(t)], color='blue', linewidth=1.5, linestyle="--") # A dotted line at the point of interest.

plt.scatter([t,],[y_sin(t)], 50, color="blue")

plt.annotate(r'$sin(\frac{\pi}{2})= 1$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+30, +30), textcoords='offset points', fontsize=12,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.legend(loc="lower left")

plt.show()
```

An interesting **exercise** for later would be to make an interactive/animated plot of the sine and cosine graphs showing values of both graphs over a particular range.

### Subplots

We can also add subplots, to juxtapose two graphs, add labels to the x and y axis and add figures for the data that allow us to customise the plot in a variety of ways. Although they are always arranged in a grid structure, which is defined in the parameters of the method the more powerful alternative is the [gridspec](https://matplotlib.org/users/gridspec.html) method.

```python
# This creates a subplot of height 2 (plots) and width 1
# It also sets the first subplot to be active
plt.subplot(2,1,1)

plt.plot(x, y_sin(x))
plt.title("Sine")

# Set the second graph active
plt.subplot(2,1,2)

plt.plot(x, y_cos(x))
plt.title("Cosine")

plt.show()
```

Again there is much more to be explored of the subplot feature which can be found in the [documentation](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot) here. 



### Figures

The figure itself is a heirarchy of python objects, the complete dissection of the figure can be found [here](https://matplotlib.org/examples/showcase/anatomy.html). ![HELLO](anatomy.7d033ebbfbc8.PNG) but is shown above to illustrate the heirarchy, I have saved a local copy of the picture but it is itself written in matplotlib which is pretty cool!

A figure refers to the whole window, you will notice that above we did not create a figure, that is because a figure is automatically created when plot is called, so all of the above plots are using the default settings, figures, and indeed, plots themselves provide a vast array of customisation.


```python
# Creating a figure, the default figure number is 1, making is 10in x 10in at 80DPI
plt.figure(figsize=(6,4), dpi=80)

plt.subplot(1,1,1)

# Plot sine with blue line with width of 2
plt.plot(x, y_sin(x), color="blue", linewidth=2.0, linestyle="-")

#Plot cosine with a green line with a width of 2
plt.plot(x, y_cos(x), color="green", linewidth=2.0, linestyle="-")

#Set x limits
plt.xlim(0, 2*np.pi)

plt.ylim(-1,1)
```
