## Exercise 1: Drawing some lines

I am going to try and draw a basic line in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np 

def line(x):
    y = 3*x
    return y

x = np.arange(0,50,0.1)

plt.title("Draw a line!")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x, line(x))

plt.show()
```

The next graph is a piecewise function plotted over an xrange of 1.0 to 3.0 and with two straight lines, one with a gradient of 2 and the other with a gradient of -3.

```python
# Creating an empty array to fill with the values from our list, could also be from a text value.
x = np.zeros(3)
y = np.zeros(3)
x_y = np.array([[1.0,2.0],[2.0,4.0],[3.0,1.0]])

# Get the x values and y values
# Iterate over the x_y array and grab each coordinate.
i = 0
for row in x_y:
    x[i] = row[0]
    y[i] = row[1]
    i += 1

# Change the x and y limits
plt.xlim(1.,3.)
plt.ylim(1.0,4.0)

plt.xlabel("x")
plt.ylabel("y")

# Plot the x and y values
plt.plot(x,y)
plt.show()

```

## Exercise 2: Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016

Here we are given the values of the publicly listed company Alphabet for their stock value through 5 days of trading, I will use a list, however it would be useful to know how to load data from a csv for example, or an excel spreadsheet. The data looks like this:

|Date|Open|High|Low|Close|
|----|----|----|---|-----|
|03-10-16|774.25|776.065002|769.5|772.559998|
|04-10-16|776.030029|778.710022|772.890015|776.429993|
|05-10-16|779.309998|782.070007|775.650024|776.469971|
|06-10-16|779|780.47998|775.539978|776.859985|
|07-10-16|779.659973|779.659973|770.75|775.080017|

```python
# Create arrays
dates = np.array(["03-10-16","04-10-16","05-10-16","06-10-16","07-10-16"])
valopen = np.array([774.25,776.030029,779.309998,779,779.659973])
high = np.array([776.065002,778.710022,782.070007,780.47998,779.659973])
low = np.array([769.5,772.890015,775.650024,775.539978,770.75])
close = np.array([772.559998,776.429993,776.469971,776.859985,775.080017])

# Set xticks to be the dates
plt.xticks = dates

plt.plot(dates, valopen, color="blue", label="open")
plt.plot(dates, high, color="red", label="high")
plt.plot(dates, low, color="cyan", label="low")
plt.plot(dates, close, color="orange", label="close")

plt.title("Alphabet Stock Value")
plt.legend()
plt.show()
```

## Bar charts

Bar charts are a great way to display information, I will be starting with some example code. I didn't really have to look anything up here because the code was fairly self explanatory, howeve I have written some comments that explain it in a little more detail.

On method that I was unfamiliar with was `zip()`, which as far as I can see combines two arrays which is quite neat!

```python
n = 10
X = np.arange(n)
# Generates 
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

# Take all of the x,y of the blue bars
for x, y in zip(X, Y1):
    plt.text(x, y + 0.05, '%.2f' % y, ha='center', va='bottom')

# Take all of the x,y of the red bars
for x, y in zip(X, Y2):
    plt.text(x, -y-0.2, '%.2f' % y, ha='center', va='bottom')

plt.ylim(-1.25, +1.25)
```

## Pie charts

The next exercise to make a grayscale gradient pie chart, starting from some code

    Z = np.random.uniform(0, 1, 20)
    plt.pie(Z)

As it stands, the [pie chart](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html) that is made has multiple colors, I want to make it grayscale with all even sizes. A pie chart takes values between 0 and 1 to produce the sizes of the wedges. By default the wedges are colored by the cycling through [matplotlib color](https://matplotlib.org/api/colors_api.html) args.

To get this I had to make an array of the color values

```python
import matplotlib.colors as color

# Specify the number of wedges, evenly divide the segments.
n=20
Z = np.ndarray(n)
Z.fill(1/n)

# Explode all of them a lilbit
ex = np.ndarray(n)
ex.fill(0.05)
# Explode the first one a bit more
ex[0] = 0.1

# Make a lil color thing
c = np.ndarray((n,4))
c.fill(0)
inc = 0
for col in c:
    add = 1./n
    col[3] = add * inc
    inc += 1


#print(c)
pie = plt.pie(Z, explode=ex, startangle=0, colors=c)

# For each patch in the pie chart set the edgecolor and the linewidth
for patch in pie[0]:
    patch.set_edgecolor('black')

plt.show()
```
