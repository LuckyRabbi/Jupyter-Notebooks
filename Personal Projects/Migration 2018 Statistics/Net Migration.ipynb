# New Zealand Migration Statistics

The goal of this document is to visualise the migration of New Zealand, these data are interesting for a variety of reasons, not least due to the unique nature of this country, with a low level of corruption, a stable economy and a unique geographic location.

I have downloaded the migration statistics from [here](https://www.stats.govt.nz/large-datasets/csv-files-for-download/) and I will be playing with them in order to gain a better understanding of Pandas, the library that I am learning for data analysis.

The first goal that I have is to visualise the net migration from all countries to New Zealand throughout time, this seems like the simplest place to start. 


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# import the csv as a DataFrame
mig = pd.read_csv("itm-oct18-plt-key-series.csv")
```

## The Data

Now that I have the dataset read and saved as a variable I need to figure out what I want to do with it, something that would be interesting would be to get all of the NZ and non-NZ immigration to see how they compare to one another.

I know that the first column are the xlabels that I am wanting, it looks like the dates are from November 2008 to Novermber 2018, that's pretty cool. I think the best way to proceed is to get rid of all the values except for net migration

```python
# I'll need to clean up the dataset first.
# Exclude all entries that aren't net migration, aren't All Countries and sort by the time value
mig_net = mig[(mig['Direction'] == "Net migration") & (mig['Country'] != "Australia")].sort_values("Period")

# Now we only have the net migration statistics and they are sorted by year
mig_nonNZ = mig_net[mig_net['Citizenship'] == "non-NZ"]
mig_NZ = mig_net[mig_net['Citizenship'] == "NZ"]

# Get the dates of the values
dates = pd.to_datetime(mig_nonNZ['Period'].values, format="%Y%m")

# Get the values that correspond to the dates
non_nz = mig_nonNZ['Count'].values
nz = mig_NZ['Count'].values
```

## The Graph

Now that the data is all in order, after a bit of learning, it's time to get the graph made. I started wanting two sperate graphs, this spurred a great amount of learning about the subplots feature of matplotlib.

The cool thing about having filtering the data is that it is now suitable for

### The Bar Graph



```python
fig, (ax1,ax2) = plt.subplots(1, 2, sharex = True, sharey = True, figsize = (16,4))

# Configure the axes, there must be a better way to do this!
ax1.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')
ax1.spines['left'].set_color('none')
ax1.spines['bottom'].set_position(('data',0))

ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')
ax2.spines['left'].set_color('none')
ax2.spines['bottom'].set_position(('data',0))

# Check it
# print(ax1.spines)

# PLot the bars
ax1.bar(dates, non_nz, width=9)
ax2.bar(dates, nz, width=9)

plt.show()
```

```python
# Importing the dates package from matplotlib
import matplotlib.dates as mdates

# Get rid of all the figures
plt.close('all')

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
yearsFmt = mdates.DateFormatter('%Y')

# Make a new figure
fig, ax = plt.subplots(figsize=(16,10))

# format the ticks
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(yearsFmt)
ax.xaxis.set_minor_locator(months)
# Format the dates
fig.autofmt_xdate()

ax.grid(True)

# Get rid of the spines, I think that they are ugly
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_position(('data',0))

ax.set_title("New Zealand Net Migration")

# Create the 
b_non_nz = ax.bar(dates, non_nz, width=20, label="Non-New Zealand Citizenship")
b_nz = ax.bar(dates, nz, width=20, label="New Zealand Citizenship")

for bar in zip(b_non_nz, b_nz):
    bar[0].set_alpha(0.5)
    bar[1].set_alpha(0.5)

ax.legend()    

plt.show()
```
