I am taking a look at the GHG emissions data from 1990 to 2014 and am going to plot them in matplotlib using pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ghg = pd.read_csv('ghg2016.csv', index_col=None)

# Set ylimits
plt.ylim(0, 90)

plt.title('NZ GHG Emissions 1990-2014')
plt.plot(ghg['December year'].values, ghg['Total Emissions'].values)
plt.plot(ghg['December year'].values, ghg['Tonnes (million) C02 equivalent '].values)

```
